Vytvorili sme si funkciu na predspracovanie v ktorej sme si spojili v≈°etky pr√≠spevky do jedn√©ho, potom sme odstr√°nili hashtagy, oznaƒçenia, emoji, nealfabeticke znaky, Unicode kontroln√© znaky, nov√© riadky a tabul√°tory. N√°sledne sme rozdelili text na jednotliv√© slov√° (tokeniz√°cia), odstr√°nila stopwords (a, the, on...) a nakoniec urobili lematiz√°ciu, ƒço je ponechanie iba kore≈àa slova

```python
def preprocess_text_h1(text):
    if isinstance(text, list):
        text = ' '.join(text)
        
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    # Odstr√°nenie u≈æ√≠vateƒæsk√Ωch oznaƒçen√≠ (@) a hashtagov (#)
    text = re.sub(r'@\w+|#\w+', '', text)
    # Odstr√°nenie emoji
    text = emoji.replace_emoji(text, replace='')
    #Odstr√°nenie nealfabetick√Ωch znakov
    text = re.sub(r"[^a-zA-Z0-9 ']", '', text)
     # Odstr√°nenie Unicode kontroln√Ωch znakov
    text = re.sub(r'[\u0000-\u001F\u0080-\u009F]', '', text)
    # Odstr√°nenie nov√Ωch riadkov a tabul√°torov
    text = text.replace('\n', ' ').replace('\r', ' ').replace('\t', ' ')
    
    # Tokeniz√°cia
    tokens = word_tokenize(text)
    
    # Odstr√°nenie stop slov a konvertovanie na mal√© p√≠smen√°
    stop_words = set(stopwords.words('english'))
    tokens = [token.lower() for token in tokens if token.lower() not in stop_words and token.isalpha()]
    
    # Lemmatiz√°cia
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token) for token in tokens]
    
    return tokens
```

```python
hypotez1['tokens'] = hypotez1['text'].apply(preprocess_text_h1)
```

ƒéalej sme sa chceli pozrie≈• na outlierov

```python
def identify_outliers(a):
    low = a.quantile(0.25) - 1.5 * stats.iqr(a)
    big = a.quantile(0.75) + 1.5 * stats.iqr(a)
    
    return a[(a > big) | (a < low)]
```

Zauj√≠malo ma ak√© s√∫ najƒçastej≈°ie slov√° v slovnej z√°sobe, a≈æ som si ist√°, ≈æe s√∫ zmyslupln√© a v anglickom jazyku (a≈æ som si ist√°, ≈æe aj predspracovanie a ƒçistenie datasetu by malo by≈• spr√°vne) a teda analyzujem slova v spravnom tvare.

```python
# Inicializ√°cia Counter objektu pre poƒç√≠tanie v√Ωskytov slov
word_counts = Counter()

# Prech√°dzanie ka≈æd√©ho zoznamu v hypotez1['tokens'] a poƒç√≠tanie slov
for token_list in hypotez1['tokens']:
    word_counts.update(token_list)

# Z√≠skanie najƒçastej≈°√≠ch 20 slov
most_common_words = word_counts.most_common(20)

# Rozdelenie zoznamu dvoj√≠c na dva zoznamy pre vizualiz√°ciu
words, counts = zip(*most_common_words)

# Vytvorenie histogramu
plt.figure(figsize=(10, 6))
plt.bar(words, counts)
plt.title('Najƒçastej≈°ie slov√°')
plt.ylabel('Poƒçetnos≈•')
plt.xticks(rotation=45)
plt.show()
```

Najƒçastej≈°ie slov√° vyzeraj√∫ v poriadku

Dalej som si chcela zisti≈•, ƒçi existuje nejak√Ω vz≈•ah medzi vekom a slovnou z√°sobou pomocou korelaƒçnej anal√Ωzy

```python
groups = [group['vocab_size'].dropna() for name, group in hypotez1.groupby('age_group')]

# Testovanie norm√°lnosti pre ka≈æd√∫ skupinu
normality = [stats.shapiro(group) for group in groups]

# V√Ωpis v√Ωsledkov Shapiro-Wilkovho testu
print("V√Ωsledky Shapiro-Wilkovho testu pre norm√°lnos≈•:")
for i, (stat, p) in enumerate(normality):
    print(f"Vekov√° skupina {i}: Statistika={stat}, p-hodnota={p}")

# Kontrola, ƒçi v≈°etky skupiny pre≈°li testom norm√°lnosti (p > 0.05)
if all(p > 0.05 for _, p in normality):
    print("\nV≈°etky skupiny maj√∫ norm√°lne rozdelenie.")

    # Vykonanie Leveneho testu na overenie homogenity varianci√≠
    levene_stat, levene_p = stats.levene(*groups)
    print(f"Leveneho test: Statistika={levene_stat}, p-hodnota={levene_p}")

    if levene_p > 0.05:
        # Variancie s√∫ rovnak√©, m√¥≈æeme pou≈æi≈• ANOVA
        print("Variancie medzi skupinami s√∫ rovnak√©. Vykon√°vam ANOVA test.")
        f_value, p_value = stats.f_oneway(*groups)
        print(f'ANOVA test: F={f_value}, p={p_value}')
    else:
        # Variancie nie s√∫ rovnak√©, m√¥≈æeme pou≈æi≈• Welchov ANOVA
        print("Variancie medzi skupinami nie s√∫ rovnak√©. Vykon√°vam Welchov ANOVA test.")
        kruskal_stat, kruskal_p = stats.kruskal(*groups)
        print(f'Kruskal-Wallisov test: H={kruskal_stat}, p={kruskal_p}')
else:
    print("\nNie v≈°etky skupiny maj√∫ norm√°lne rozdelenie. Vykon√°vam Kruskal-Wallisov test.")
    kruskal_stat, kruskal_p = stats.kruskal(*groups)
    print(f'Kruskal-Wallisov test: H={kruskal_stat}, p={kruskal_p}')
```

Z testov m√¥≈æeme vidie≈•, ≈æe v≈°etky skupiny nemaj√∫ norm√°lne rozdelenie (pre norm√°lne hodnoty musi by≈• p>0.05). Tak≈æe sme i≈°li vykona≈• Kruskal-Wallisov test. Jeho hodnota p<0.05, ƒço n√°m indikuje, ≈æe existuj√∫ v√Ωznamn√© rozdiely medzi vekov√Ωmi skupinami.

Uprav√≠me si pr√≠spevky (text) a zist√≠m jednotliv√© dl≈æky viet

```python
def preprocess_text_h2(text):
    if isinstance(text, list):
        text = ' '.join(text)
        
    sentences = sent_tokenize(text)#tokeniyacia na vety
    sentence_lengths = []  # na ukladanie dƒ∫≈æok viet
    for sentence in sentences:
        sentence = re.sub(r'http\S+|www\S+|https\S+', '', sentence, flags=re.MULTILINE)
        # Odstr√°nenie u≈æ√≠vateƒæsk√Ωch oznaƒçen√≠ (@) a hashtagov (#)
        sentence = re.sub(r'@\w+|#\w+', '', sentence)
        # Odstr√°nenie emoji
        sentence = emoji.replace_emoji(sentence, replace='')
        #Odstr√°nenie nealfabetick√Ωch znakov
        sentence = re.sub(r"[^a-zA-Z0-9 ']", '', sentence)
         # Odstr√°nenie Unicode kontroln√Ωch znakov
        sentence = re.sub(r'[\u0000-\u001F\u0080-\u009F]', '', sentence)
        # Odstr√°nenie nov√Ωch riadkov a tabul√°torov
        sentence = sentence.replace('\n', ' ').replace('\r', ' ').replace('\t', ' ')
        tokens = word_tokenize(sentence)
         # Odstr√°nenie stop slov a konvertovanie na mal√© p√≠smen√°
        stop_words = set(stopwords.words('english'))
        tokens = [token.lower() for token in tokens if token.lower() not in stop_words and token.isalpha()]
        sentence_lengths.append(len(tokens))  # Ulo≈æenie dƒ∫≈æky upravenej vety
    return sentence_lengths

```

####  H3: ≈Ωeny pou≈æ√≠vaj√∫ viac slov√° s pozit√≠vnym citov√Ωm sentimentom ako mu≈æi.

Upravili sme si pr√≠spevky (text) a n√°sledne zistili pomer pozit√≠vnych slov ku v≈°etk√Ωm slov√°m pre ka≈æd√©ho ƒçloveka

```python
sid = SentimentIntensityAnalyzer()

def preprocess_text_h3(text):
    if isinstance(text, list):
        text = ' '.join(text)
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    # Odstr√°nenie u≈æ√≠vateƒæsk√Ωch oznaƒçen√≠ (@) a hashtagov (#)
    text = re.sub(r'@\w+|#\w+', '', text)
    # Odstr√°nenie emoji
    text = emoji.replace_emoji(text, replace='')
    #Odstr√°nenie nealfabetick√Ωch znakov
    text = re.sub(r"[^a-zA-Z0-9 ']", '', text)
     # Odstr√°nenie Unicode kontroln√Ωch znakov
    text = re.sub(r'[\u0000-\u001F\u0080-\u009F]', '', text)
    # Odstr√°nenie nov√Ωch riadkov a tabul√°torov
    text = text.replace('\n', ' ').replace('\r', ' ').replace('\t', ' ')
    # Tokeniz√°cia textu na slov√°
    words = word_tokenize(text)
     # Odstr√°nenie stop slov a konvertovanie na mal√© p√≠smen√°
    stop_words = set(stopwords.words('english'))
    words = [word.lower() for word in words if word.lower() not in stop_words and word.isalpha()]
    # Poƒçet slov s pozit√≠vnym sentimentom
    positive_words = sum([sid.polarity_scores(word)['compound'] > 0.1 for word in words])
    # Celkov√Ω poƒçet slov
    total_words = len(words)
    # Vr√°ti≈• pomer pozit√≠vnych slov k celkov√©mu poƒçtu slov
    return positive_words / total_words if total_words > 0 else 0
```

```python
hypotez3['positive_words_ratio'] = hypotez3['text'].apply(preprocess_text_h3)
```
