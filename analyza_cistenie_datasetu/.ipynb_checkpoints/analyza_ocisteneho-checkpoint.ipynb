```python
#inportovanie potrebnych kniznic
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import jsonlines
import json
from bs4 import BeautifulSoup
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from datetime import datetime
import matplotlib.pyplot as plt
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
from scipy.stats import shapiro
from scipy.stats import kruskal
from scipy.stats import ttest_ind, f_oneway
from scipy.stats import pearsonr, spearmanr
from nltk.tokenize import sent_tokenize, word_tokenize
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

Na캜칤tanie potrebn칠ho datasetu

Ur캜ili sme si vekov칠 skupiny a zistilipo캜et 쬿d칤 v nich

```python
#zistenie veku
current_year = datetime.now().year
analysis['age'] = current_year - analysis['birthyear']
#zaradenie do vekovej kategorie
bins = [0, 30, 40, 50, 60, 70, 80, 100]
labels = ['0-29', '30-39', '40-49', '50-59', '60-69', '70-79', '80+']
analysis['age_group'] = pd.cut(analysis['age'], bins=bins, labels=labels, right=False)
#zistenie poctu zaznamov pre ka쬯u kategoriu
age_group_counts = analysis['age_group'].value_counts().sort_index()

print("Po캜et 쬿d칤 v datasete pod쬬 vekov칳ch skup칤n:")
print(age_group_counts)
```

Pre lep코iu predstavu sme si to vizuovali v grafe

```python
plt.figure(figsize=[10,6])
age_group_counts.plot(kind='bar', color='pink') 
plt.title('Po캜et 쬿d칤 v datasete pod쬬 vekov칳ch skup칤n')
plt.xlabel('Vekov치 skupina') 
plt.ylabel('Po캜et 쬿d칤')
plt.xticks(rotation=45) 
plt.grid(axis='y', linestyle='--', alpha=0.7) 
plt.show() 
```

Ako m칪쬰me vidie콘 na grafe, tak najviac 쬿d칤 je vo veku 30-39 rokov v mojom datasete

N치sledne sme si pozreli ko쬶o mu쬺v a 쬴en je v jednotliv칳ch vekov칳ch skupin치ch

```python
gender_age_distribution = analysis.groupby(['gender', 'age_group']).size().unstack(fill_value=0)

print("Po캜et 칰dajov pod쬬 pohlavia a vekov칳ch skup칤n:")
print(gender_age_distribution)
```

N치sledne sme si to aj vizualizovali

```python
my_palette = sns.color_palette("pastel")
gender_age_distribution.plot(kind='bar', figsize=(12, 8), color=my_palette)

plt.title('Distrib칰cia po캜tu 칰dajov pod쬬 pohlavia a vekov칳ch skup칤n')
plt.xlabel('Vekov치 skupina')
plt.ylabel('Po캜et 칰dajov')
plt.legend(title='Pohlavie')
plt.xticks(rotation=45)
plt.show()
```

Na grafe vy코코ie m칪쬰me vidie콘, 쬰 najviac 쬴en aj mu쬺v je vo vekov칳ch kateg칩ri치ch 30-39 a 40-49, ktor칠 s칰 najpo캜etnej코ie skupiny v datasete; toto zistenie je relat칤vne o캜ak치van칠. Vzh쬬dom na to, 쬰 mu쬺v je viac, s칰 v칳raznej코ie zast칰pen칤 vo v코etk칳ch vekov칳ch kateg칩ri치ch. Preto sme sa rozhodli znormalizova콘 vekov칠 kateg칩rie, aby sme eliminovali vplyv nerovnomernej distrib칰cie na v칳sledky. Z normalizovan칠ho grafu ni쮄멸e je zrejm칠, 쬰 mu쬴 s칰 viac zast칰pen칤 v star코칤ch vekov칳ch kateg칩ri치ch a v kateg칩rii 30-39, zatia 캜o 쬰ny prevl치daj칰 v mlad코칤ch vekov칳ch skupin치ch s v칳nimkou kateg칩rie 30-39.

```python
# Najprv spo캜칤tame po캜et z치znamov pod쬬 pohlavia a vekovej skupiny
gender_age_distribution = analysis.groupby(['gender', 'age_group']).size().unstack(fill_value=0)

# Zist칤me celkov칠 po캜ty pre ka쬯칰 vekov칰 skupinu v celom datasete
total_per_age_group = analysis.groupby('age_group').size()

# Normalizujeme 칰daje pod쬬 celkov칠ho po캜tu v ka쬯ej vekovej skupine
normalized_distribution = gender_age_distribution.div(total_per_age_group, axis=1) * 100

# Konvertujeme kategorick칠 typy na string, aby sa predi코lo probl칠mom s kategorick칳m indexom
normalized_distribution.columns = normalized_distribution.columns.astype(str)
normalized_distribution = normalized_distribution.reset_index()

# Priprav칤me 칰daje pre vizualiz치ciu
data = pd.melt(normalized_distribution, id_vars='gender', value_vars=normalized_distribution.columns[1:], var_name='age_group', value_name='percent')

# Vizualiz치cia normalizovan칳ch 칰dajov
plt.figure(figsize=(12, 8))
sns.barplot(data=data, x='gender', y='percent', hue='age_group', palette='pastel')

plt.title('Normalizovan치 percentu치lna distrib칰cia pod쬬 vekov칳ch skup칤n a pohlavia')
plt.ylabel('Percento (%)')
plt.xlabel('Pohlavie')
plt.legend(title='Vekov치 skupina')
plt.xticks(rotation=0)
plt.show()
```

```python
# Najprv spo캜칤tame po캜et z치znamov pod쬬 pohlavia a vekovej skupiny
gender_age_distribution = analysis.groupby(['gender', 'age_group']).size().unstack(fill_value=0)

# Priprav칤me 칰daje pre vizualiz치ciu a preusporiadame d치ta na potrebn칳 form치t
gender_age_distribution = gender_age_distribution.T.reset_index()
melted_data = pd.melt(gender_age_distribution, id_vars='age_group', value_vars=['female', 'male'], var_name='gender', value_name='count')

# Vizualiz치cia 칰dajov
plt.figure(figsize=(12, 8))
sns.barplot(data=melted_data, x='age_group', y='count', hue='gender', palette='pastel')
plt.title('Absol칰tne po캜ty pohlav칤 v r칪znych vekov칳ch skupin치ch')
plt.ylabel('Po캜et 쬿d칤')
plt.xlabel('Vekov치 skupina')
plt.legend(title='Pohlavie')
plt.xticks(rotation=45)
plt.show()
```

Na grafe m칪쬰me vidie콘, 쬰 vo v코etk칳ch vekov칳ch kategoriach prevl치daj칰 mu쬴. Je to predpokladan칠, ked쬰 mu쬿 tvoria v칛캜코iu 캜as콘 datasetu. Preto sme si sk칰sili znormalizova콘 pohlavie, aby sme minimalizovali vply distrib칰cie na v칳sledok a aby sme lep코ie videli po캜etnos콘 pohlav칤 medzi vekov칳mi kategoriami. Na grafe ni쬽ie u m칪쬰me vidie콘, 쬰 rozdieli medzi pohlaviami nie s칰 a tak zna캜n칠, v niektor칳ch kategoriach prevl치daj칰 mu쬴, v niektor칳ch 쬰ny.

```python
# Najprv spo캜칤tame po캜et z치znamov pod쬬 pohlavia a vekovej skupiny
gender_age_distribution = analysis.groupby(['gender', 'age_group']).size().unstack(fill_value=0)

# Zist칤me celkov칠 po캜ty pre ka쬯칠 pohlavie
total_per_gender = gender_age_distribution.sum(axis=1)

# Normalizujeme 칰daje pod쬬 celkov칠ho po캜tu mu쬺v a 쬴en
normalized_distribution = gender_age_distribution.div(total_per_gender, axis=0) * 100

# Priprav칤me 칰daje pre vizualiz치ciu a preusporiadame d치ta na potrebn칳 form치t
normalized_distribution = normalized_distribution.T.reset_index()
melted_data = pd.melt(normalized_distribution, id_vars='age_group', value_vars=['female', 'male'], var_name='gender', value_name='percent')

# Vizualiz치cia normalizovan칳ch 칰dajov
plt.figure(figsize=(12, 8))
sns.barplot(data=melted_data, x='age_group', y='percent', hue='gender', palette='pastel')

plt.title('Percentu치lne zast칰penie pohlav칤 v r칪znych vekov칳ch skupin치ch')
plt.ylabel('Percento (%)')
plt.xlabel('Vekov치 skupina')
plt.legend(title='Pohlavie')
plt.xticks(rotation=45)
plt.show()
```

Preto sme sa to rozhodli zobrazi콘 aj percenntu치lne, aby sme videla naozajst칠 zast칰penie pohlav칤 v kategoriach bez oh쬬du na ich mno쬽tvo v datasete. Predelili sme si po캜ty pohlav칤 v jednotliv칳ch kategoriach po캜tom celkov칠ho po캜tu dan칠ho pohlavia v datasete. Ako m칪쬰te vidie콘 na grafe ni쮄멸e, tak rozdieli medzi pohlaviami v jednotliv칳ch kategoriach sa skoro vytratili.

```python
# Vypo캜칤tame percentu치lne zast칰penie
post_gender_distribution['female_percentage'] = (post_gender_distribution['female'] / gender_counts['female']) * 100
post_gender_distribution['male_percentage'] = (post_gender_distribution['male'] / gender_counts['male']) * 100
print(post_gender_distribution)

# Pretransformujeme d치ta pre vizualiz치ciu
post_gender_distribution.reset_index(inplace=True)
post_gender_distribution_wide2 = post_gender_distribution.melt(id_vars='post_category', value_vars=['female_percentage', 'male_percentage'], var_name='gender', value_name='percentage')

# Vytvorte graf
plt.figure(figsize=(10, 6))
sns.barplot(x='post_category', y='percentage', hue='gender', data=post_gender_distribution_wide2, palette=custom_palette )
plt.title('Percentu치lne zast칰penie pr칤spevkov pod쬬 kateg칩ri칤 a pohlavia')
plt.xlabel('Kateg칩ria po캜tu pr칤spevkov')
plt.ylabel('Percento pr칤spevkov')
plt.legend(title='Pohlavie')
plt.xticks(rotation=45)
plt.show()
```

V obidvoch kategoriach s po캜tom z치znamov dominuj칰 mu쬴

Pozreli sme si po캜ty pr칤spevkov na z치klade vekovej skupiny

```python
#vytvorenie tabulky
post_age_distribution = pd.crosstab(analysis['post_category'], analysis['age_group'])

print("Rozdelenie po캜tu pr칤spevkov pod쬬 kateg칩ri칤 a vekov칳ch skup칤n:")
print(post_age_distribution)
```

```python
my_palette = sns.color_palette("pastel")
post_age_distribution.plot(kind='bar', figsize=(14, 7), color=my_palette)
plt.title('Rozdelenie po캜tu pr칤spevkov pod쬬 kateg칩ri칤 a vekov칳ch skup칤n')
plt.xlabel('Kateg칩ria po캜tu pr칤spevkov')
plt.ylabel('Po캜et pr칤spevkov')
plt.xticks(rotation=45)
plt.legend(title='Vekov치 kateg칩ria')
plt.show()
```

Ako m칪쬰me vidie콘 na grafe vy코코ie, tak najviac prevl치dza vekov치 kategoria 30-39 a 40-49, 캜o maj칰 najv칛캜ie zast칰penie v datasete. Toto je nie캜o, 캜o o캜ak치vame, ked쬰 vekov칠 kategorie nie s칰 rovnomerne z치st칰pen칠. Preto sme znova sa rozhodli pre percentu치lne zobrazenie aj v tomto pr칤pade. Ako m칪쬰me vidie콘 na grafe pod tymto textom, tak rozdiely medzi vekov칳mi kategoriami sa zna캜ne zmen코ili. Menej pr칤spevkov maj칰 mlad코ie alebo star코ie vekov칠 kategorie. Mlad코칤 쬿dia maj칰 sa viac zaraduj칰 do kategorie s menej pr칤spevkami a star코칤 naopak do kategorie s viacer칳mi pr칤spevkami.

```python
post_age_distribution.columns = post_age_distribution.columns.astype(str) 

for column in post_age_distribution.columns:
    # Vytvor칤me nov칳 st컄pec pre ka쬯칰 vekov칰 skupinu
    age_group = column.split('_')[0]  # Oddel칤me n치zov vekovej skupiny
    if age_group in age_group_counts.index:
        total_count = age_group_counts[age_group]
        post_age_distribution[column + '_percentage'] = (post_age_distribution[column] / total_count) * 100

# Preform치tovanie d치t pre vizualiz치ciu
post_age_distribution2 = post_age_distribution.reset_index().melt(id_vars='post_category', value_vars=[col for col in post_age_distribution.columns if 'percentage' in col], var_name='age_group', value_name='percentage')

pastel_palette = sns.color_palette("pastel")
# Vytvorte graf
plt.figure(figsize=(14, 7))
sns.barplot(x='post_category', y='percentage', hue='age_group', data=post_age_distribution2, palette=pastel_palette)
plt.title('Percentu치lne zast칰penie pr칤spevkov pod쬬 kateg칩ri칤 a vekov칳ch skup칤n')
plt.xlabel('Kateg칩ria po캜tu pr칤spevkov')
plt.ylabel('Percento pr칤spevkov')
plt.legend(title='Vekov치 kateg칩ria')
plt.xticks(rotation=45)
plt.show()
```

V z치znamoch s po캜tom pr칤spevkov 301-400 dominuje vekov치 kategoria 30-39 a v z치znamoch s po캜tom pr칤spevkov 401-500 dominuje vekov치 kategoria 40-49
